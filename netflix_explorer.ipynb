**Project Description:**  
This notebook analyzes Netflix's global catalogue using pandas. It covers content distribution across countries, content types, genre popularity, director contributions, and release trends.

**Dataset:** [Netflix Titles on Kaggle](https://www.kaggle.com/datasets/shivamb/netflix-shows)


```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", 25)
```

```python
netflix = pd.read_csv('netflix_titles.csv')
```

## Cleaning Data

```python
netflix.fillna(np.nan)
```

```python
netflix.isna().sum()
#Therefore there are alot of missing values which need to be dropped
```

```python
netflix.dropna(inplace=True)
```

```python
#We finished missing value Treatment now we need to find if there are any duplicates
netflix.duplicated().sum()
#Therefore no duplicated values
```

## Q1. Top 10 Countries with Most Shows

```python
top_10_countries = netflix['country'].value_counts().head(10).index.to_list()
print(top_10_countries)
```

## Q2. Count of movies vs tv shows

```python
netflix
```

```python
netflix['type'].value_counts()
```

## Q3. Trend of content release over years

```python
release_trend = pd.DataFrame(netflix['release_year'].value_counts())
release_trend = release_trend.rename(columns={'count':'total_titles'})
print(release_trend)
```

## Q4. Most frequent genres (based on listed_in)

```python
genre_list = netflix['listed_in'].str.split(', ')

from itertools import chain

new_genre_list = list(chain.from_iterable(genre_list))
popular_genre = pd.Series(new_genre_list).value_counts()
print(popular_genre)
```

## Q5. Top directors by volume

```python
netflix
directors = netflix['director'].str.split(', ')
directors = list(chain.from_iterable(directors))
top_10_directors = pd.Series(directors).value_counts().head(10)
top_10_directors
```
